**Ques 2.** Explain the difference between mutable and immutable data types with examples.

**Ans.** Data types in Python are categorized into mutable and immutable data types. Mutable data types are those whose values can be changed, whereas immutable data types are ones in which the values can’t be changed. In this article, we will discuss the difference between mutable and immutable in Python.

 Example:- Mutable Data Types:

**Lists:** Lists are ordered, changeable (mutable) collections of items. They can
contain different data types.

    my_list =
    my_list.append(4)  # Modifying the list
    print(my_list)  # Output:

**Dictonaries:** Dictionaries are unordered collections of key-value pairs. They are mutable, allowing you to add, remove, or modify key-value pairs.

    my_dict = {"name": "Alice", "age": 30}
    my_dict["city"] = "New York"  # Modifying the dictionary
    print(my_dict)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}

**Sets:** Sets are unordered collections of unique items. They are mutable, allowing you to add or remove elements

    my_set = {1, 2, 3}
    my_set.add(4)  # Modifying the set
    print(my_set)  # Output: {1, 2, 3, 4}

**Immutable Data types:**

**Strings:** Strings are sequences of characters. Once created, their characters cannot be changed. Instead, operations on strings create new strings.
    my_string = "hello"
    # my_string = "H"  # This would cause an error
    new_string = "H" + my_string[1:] # Creating a new string
    print(new_string) # Output: Hello

**Tuples:** Tuples are ordered collections of items, similar to lists, but they are immutable. Operations that would modify a list, like adding or removing elements, are not allowed on tuples.

    my_tuple = (1, 2, 3)
    # my_tuple = 4 # This would cause an error
    new_tuple = my_tuple + (4,) # Creating a new tuple
    print(new_tuple) # Output: (1, 2, 3, 4)

**Ques.5**  Why might you use a set instead of a list in python?

**Ans.** You might use a set instead of a list in Python primarily for the following reasons:


**Uniqueness of Elements:** Sets inherently store only unique elements. If you

need a collection where duplicate values are not allowed, a set automatically handles this by preventing the addition of identical items. Lists, conversely, allow duplicates.

**Efficient Membership Testing (Lookup):** Sets are optimized for checking if an element exists within the collection. This is due to their underlying implementation using hash tables, which allows for average O(1) time complexity for membership tests (in operator). Lists, on the other hand, require iterating through elements, resulting in O(n) complexity for membership testing in the worst case. This makes sets significantly faster for large datasets when frequent lookups are performed.

**Mathematical Set Operations:** Sets provide built-in methods for common mathematical set operations like union, intersection, difference, and symmetric difference. These operations are highly optimized and simplify code when performing such comparisons between collections.
• Eliminating Duplicates: If you have a list with duplicate elements and need to obtain a collection of only unique elements, converting the list to a set and then back to a list (if order is not critical) is an efficient way to achieve this.

When not to use a set:


  When the order of elements is important, as sets are unordered collections and do not maintain insertion order.

 When duplicate elements are required in the collection.

**Ques. 7** How do tuples ensure data integrity in python?

**Ans.**
Tuples ensure data integrity in Python primarily through their immutability. This characteristic means that once a tuple is created, its elements cannot be changed, added, or removed.

**Preventing Accidental Modifications:**
The core benefit of immutability is that it safeguards data from unintentional alterations. If a piece of data is stored in a tuple, there's no risk of a subsequent operation or function accidentally modifying its contents. This is crucial for data that is meant to remain constant, like configuration settings, unique identifiers (e.g., user IDs, bank account numbers), or fixed historical records.

**Thread Safety:**
In multi-threaded environments, mutable data structures can lead to race conditions and data corruption if multiple threads try to modify the same data concurrently. Since tuples are immutable, they are inherently thread-safe, as their contents cannot be changed by any thread after creation.
Hashability and Dictionary Keys:
Because tuples are immutable, they are hashable. This allows them to be used as keys in dictionaries, which require hashable objects. Using tuples as dictionary keys can be valuable for representing composite keys, where a combination of values uniquely identifies an entry. The immutability of the tuple ensures that the key remains consistent and can be reliably hashed.

**Ques.8** What is a hash table, and how does it relate to dictionaries in python ?

**Ans.** A hash table is a data structure that implements an associative array, also known as a dictionary or map. It stores key-value pairs by using a hash function to compute an index into an array of buckets or slots, from which the desired value can be retrieved. This allows for efficient insertion, deletion, and lookup operations, typically achieving average-case time complexity of O(1).

**Relationship to Python Dictionaries:**

Python's built-in dict type is an implementation of a hash table. When you create a dictionary in Python and add key-value pairs, the following occurs:

- **Hashing the Key:** Python takes the key you provide and passes it through a hash function. This function converts the key into a fixed-size integer, known as a hash value or hash code.

- **Mapping to Index:** This hash value is then used to determine the index within an internal array (the hash table's underlying structure) where the key-value pair will be stored.

- **Storage:** The key and its corresponding value are then stored in that location.

**Ques.10** Explain why strings are immutable in python.

**Ans.** Strings are immutable in Python, meaning their content cannot be changed after creation. This design choice offers several advantages:

- **Security and Predictability:**  Immutability prevents accidental modification of string values, which is crucial for data integrity and security, especially when strings are passed between different parts of a program or used in concurrent environments. If a string were mutable, a function could modify a string passed to it, leading to unexpected behavior in other parts of the code that still reference the original string object.
- **Hashing and Dictionary Keys:** Immutability allows strings to be hashable, which is a requirement for using them as keys in dictionaries (or elements in sets). Since a string's value cannot change, its hash value remains constant, enabling efficient lookup operations.
- **Memory Optimization (Interning):** Python can optimize memory usage by "interning" strings, particularly short, commonly used strings. This means that if multiple variables refer to the same string literal (e.g., "hello"), they might all point to the same object in memory, reducing memory consumption. This optimization relies on immutability, as there's no risk of one variable modifying the shared string object and affecting others.
- **Thread Safety:** In multi-threaded environments, immutable objects are inherently thread-safe because their state cannot be changed by multiple threads simultaneously, eliminating the need for complex locking mechanisms and potential race conditions.
- **Simplicity and Readability:** Immutability simplifies reasoning about code, as the state of a string object is guaranteed to remain constant. This can lead to more predictable and easier-to-debug programs.

While strings cannot be modified in place, operations that appear to modify a string (like concatenation or replacement) actually create a new string object with the desired changes, and the variable is then re-assigned to reference this new object. The original string object remains unchanged in memory until garbage collected.

**Ques.13** How do sets handle duplicate values in python?

**Ans.**  Python sets fundamentally do not allow duplicate values. They are designed to store only unique elements.
When attempting to add a duplicate element to a set, the set will simply ignore the addition. The existing unique element will remain, and no new instance of that element will be added. This behavior is inherent to the definition of a mathematical set, which a Python set aims to model.

**Example: ***


```python
my_set = {1, 2, 3}
print(my_set)

my_set.add(2)  # Attempting to add a duplicate
print(my_set)

my_set.add(4)  # Adding a unique element
print(my_set)
```

As demonstrated, when 2 is added again, the set remains unchanged because 2 already exists. When 4 is added, it is a unique element, so it is successfully incorporated into the set.
